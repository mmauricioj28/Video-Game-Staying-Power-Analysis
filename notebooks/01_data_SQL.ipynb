```python
import pandas as pd
import sqlite3
```

```python
df_released = pd.read_csv("../data/cleaned/steam_metadata_released_clean.csv")
df_released.head()
```

```python
conn = sqlite3.connect("../database/steam_games.db")

df_released.to_sql("games_metadata", conn, if_exists="replace", index=False)

print("games_metadata table created and loaded.")

conn.close()
```

```python
conn = sqlite3.connect("../database/steam_games.db")

query = "SELECT * FROM games_metadata;"
preview = pd.read_sql(query, conn)

print(preview)

conn.close()
```
