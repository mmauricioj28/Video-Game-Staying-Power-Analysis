```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("../database/steam_games.db")
```

```python
query = """
SELECT 
    m.app_id,
    m.name,
    m.release_date,
    m.genres,
    r.review_score_desc,
    r.review_score,
    r.total_positive,
    r.total_negative,
    r.positive_ratio
FROM games_metadata m
LEFT JOIN reviews r
ON m.app_id = r.app_id
LIMIT 10;
"""

joined_preview = pd.read_sql(query, conn)
print(joined_preview)

conn.close()
```

```python
conn = sqlite3.connect("../database/steam_games.db")

query = """
SELECT 
    COUNT(*) AS total_games,
    COUNT(r.app_id) AS matched_reviews
FROM games_metadata m
LEFT JOIN reviews r
ON m.app_id = r.app_id;
"""

match_check = pd.read_sql(query, conn)
print(match_check)

conn.close()
```

```python
conn = sqlite3.connect("../database/steam_games.db")

query = """
SELECT 
    m.app_id,
    m.name
FROM games_metadata m
LEFT JOIN reviews r
ON m.app_id = r.app_id
WHERE r.app_id IS NULL;
"""

missing_reviews = pd.read_sql(query, conn)
print(missing_reviews)

conn.close()
```
