```python
import sqlite3
import pandas as pd


df_reviews_clean = pd.read_csv("../data/cleaned/steam_reviews_clean.csv")

conn = sqlite3.connect("../database/steam_games.db")

df_reviews_clean.to_sql("reviews", conn, if_exists="replace", index=False)

print("reviews table created and loaded.")

conn.close()
```
