```python
import pandas as pd
```

```python
df_check = pd.read_csv("../data/raw/steam_metadata_batch1.csv")
df_check.head()
```

```python
print(df_check.info())
print(df_check.isnull().sum())
```

```python
df_check.columns = df_check.columns.str.strip().str.lower()

df_check["release_date"] = pd.to_datetime(df_check["release_date"], errors="coerce")

df_check = df_check.drop_duplicates(subset="app_id")
print(df_check.info())
```

```python
df_released_games = df_check[df_check['coming_soon'] == False]
df_released_games = df_released_games.drop(columns = ['coming_soon', 'header_image', 'website'])
```

```python
import os
os.makedirs("../data/cleaned", exist_ok=True)

df_released_games.to_csv("../data/cleaned/steam_metadata_released_clean.csv", index=False)
df_check.to_csv("../data/cleaned/steam_metadata_all_clean.csv")
print("Cleaned file saved.")
```
