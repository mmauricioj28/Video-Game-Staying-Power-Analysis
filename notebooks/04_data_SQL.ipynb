```python
import sqlite3
import pandas as pd
```

```python
df_content_clean = pd.read_csv("..\data\cleaned\steam_content_clean.csv")

conn = sqlite3.connect('..\database\steam_games.db')

df_content_clean.to_sql('content_support', conn, if_exists='replace', index=False)

print("table loaded")

conn.close()
```

```python
conn = sqlite3.connect('..\database\steam_games.db')

query = '''
Select a.app_id,
a.latest_date,
a.tags,
b.is_free,
b.dlc_count


FROM content_support a 
LEFT JOIN games_metadata b on a.app_id = b.app_id
WHERE a.tags = "patchnotes" AND b.genres LIKE '%RPG%'

'''
print(pd.read_sql(query, conn))
```
