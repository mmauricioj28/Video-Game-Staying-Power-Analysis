```python
import sqlite3
import pandas as pd
```

```python
df_engagement_clean = pd.read_csv("../data/cleaned/steam_engagement_clean.csv")

conn = sqlite3.connect("../database/steam_games.db")

df_engagement_clean.to_sql("engagement", conn, if_exists="replace", index=False)

print("engagement table created and loaded.")

conn.close()
```

```python
conn = sqlite3.connect("../database/steam_games.db")

query = """
SELECT 
    m.app_id,
    m.name,
    m.release_date,
    e.current_players
FROM games_metadata m
LEFT JOIN engagement e
ON m.app_id = e.app_id
LIMIT 10;
"""

preview = pd.read_sql(query, conn)
print(preview)

conn.close()
```
