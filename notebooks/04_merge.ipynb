```python
import sqlite3
import pandas as pd
```

```python

conn = sqlite3.connect("../database/steam_games.db")

query = """
SELECT 
    m.app_id,
    m.name,
    m.release_date,
    m.is_free,
    m.price_usd,
    m.dlc_count,

    r.total_reviews,
    r.total_positive,
    r.total_negative,
    r.positive_ratio,

    e.current_players,

    c.latest_date,
    c.tags

FROM games_metadata m
LEFT JOIN reviews r ON m.app_id = r.app_id
LEFT JOIN engagement e ON m.app_id = e.app_id
LEFT JOIN content_support c ON m.app_id = c.app_id
"""

df = pd.read_sql(query, conn)
conn.close()

df.head()
```

```python
df.to_csv("../data/final/base_dataset.csv", index=False)
```
